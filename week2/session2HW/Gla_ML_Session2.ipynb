* Read session4_dummy_students. csv
* Create an average column using math, science, and english.
* Create a result column: Pass if average >= 40, otherwise Fail.
* Print class average.
* Print highest average and lowest average.
* Print top student or students.
* Print students with attendance below 75.
* Print students who submitted fewer than 6 assignments.
* Save final output as session4_student_report.csv.

```python
df = pd.read_csv('session4_dummy_students_data.csv')

df['Average'] = df[['math', 'science', 'english']].mean(axis=1)

df['Result'] = np.where(df['Average'] >= 40, 'Pass', 'Fail')

print(df['Average'].mean())

print(df['Average'].max())
print(df['Average'].min())

print(df[df['Average'] == df['Average'].max()])

print(df[df['attendance_percent'] < 75])

print(df[df['assignments_submitted'] < 6])
```

```python
df.to_csv('studentreport.csv', index=False)
```

*	Find the average final_project_score.
*	Find students with study_hours_per_week greater than 8.
*	Compare average marks of Mumbai students and Pune students.
*	Create a new column called performance_level.
*	Save only selected columns to a new CSV.

```python
avg_final_project_score = df['final_project_score'].mean()
print("Average final_project_score: ", avg_final_project_score)
```

```python
students = df[df['study_hours_per_week'] > 8]
print(students)
```

```python
df['marks'] = df[['math','science','english']].mean(axis=1)

avg_mumbai = df[df['city']=='Mumbai']['marks'].mean()
avg_pune = df[df['city']=='Pune']['marks'].mean()

print("Average marks of Mumbai students: ", avg_mumbai)
print("Average marks of Pune students: ", avg_pune)
```

```python
def get_performance(score):
  if score >= 90:
    return "Excellent"
  elif score >= 70:
    return "Good"
  else:
    return "Needs Improvement"

df["performance_level"] = df['final_project_score'].apply(get_performance)
print(df)
```

```python
selected_columns = ['student_id', 'name', 'performance_level']
df[selected_columns].to_csv('selected_students.csv', index=False)
```
